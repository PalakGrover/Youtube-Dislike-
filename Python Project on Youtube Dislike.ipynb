# Q3:Check for the Percentage of the missing values and drop or impute them. 

```python
# To get a boolean expression to show the missing values
filedf.isna().any()
```

```python
a = filedf.isna().sum()
print(a)
b = len(filedf)
print(b)
```

```python
c = a/b*100
# Here, "a" is the total number of null values
# "b" is the total number of rows in "filedf" dataframe
c
# Here, "c" gives the percentage of null values for each column
```

Encoded the categorical variables and found the statistical summary: https://www.kaggle.com/code/palakgro31/project-youtube-dislikes-categorical-encoding/edit

```python
# Percentage of null values
round(filedf.isna().sum()/len(filedf)*100)

# or 

null_percent= filedf.isna().sum()/len(filedf)*100
round(null_percent,2).sort_values(ascending=False)

# sort function with ascending = False to highlight the column with null-values
```

```python
# 1. We infer that the "comments" section has 0.42% of null values which can either be imputed or dropped
# 2. The better is to "impute" the "null" values as some relevant information related to rows may affect our insights
# 3. Since "comments" column is a categorical value, so better we impute it the "mode of comments" itself

filedf = filedf.apply(lambda x: x.fillna(filedf['comments'].mode()),axis=0)
filedf.head(2)
```

```python
# Cross-checking to see if the null-values are imputed with "mode" of categorical column "comments" column
filedf.isna().sum()

# Below, we verified that the null values are gone permanently, as the operation was applied on the original dataframe itself
```

# Q5 Convert datatype of column published_at from object to pandas datetime.

```python
filedf['published_at'] = filedf['published_at'].astype('datetime64')
print(filedf.info())

# Below, the datetime which was an "object" datatype has now got changed to "datetime" datatype
```

# Q6: Create a new column as 'published_month' using the column published_at (display the months only)

```python
filedf['published_month'] = pd.to_datetime(filedf['published_at']).dt.strftime('%m')
filedf.head(2)
```

# Q7: Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on..... 

```python
filedf['published_month'] = pd.to_datetime(filedf['published_month'], format='%m').dt.month_name().str.slice(stop=12)

filedf
```

# Q8: Find the number of videos published each month and arrange the months in a decreasing order based on the video count

```python
filedf.groupby('published_month').agg({'video_id':['count']}).sort_values(ascending = False,by = 'published_month')

# Below, we find that "October" month had the highest video Published count which was "4991"
```
